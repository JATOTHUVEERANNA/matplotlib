```python
import pandas as pd
```

```python
df =pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
df.head(2)
```

```python
df.tail(3)
```

```python
#no of rows and columns
print(df.shape)
```

```python
#drop problamatic rows
df = df.dropna()
```

```python
print(df.shape)
```

```python
rooms = df["Rooms"]
print(rooms)
```

```python
price = df["Price"]
print(price)
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
plt.hist(x=price,edgecolor="black", color="yellow",bins=90)
plt.xlabel("price range")
plt.ylabel("number of houses")
plt.title("histogram of price")

plt.show()
```

```python
rooms_count=df["Rooms"].value_counts()
```

```python
rooms_count=dict(rooms_count)
```

```python
rooms_count
```

```python
print(rooms_count)
```

```python
a=rooms_count.keys()
b=rooms_count.values()
```

```python
#pie
fig= plt.figure(figsize=(8,8))
plt.pie(x=b,labels = a)
plt.legend()
plt.show()
```

```python
region = df["Regionname"].value_counts()
```

```python
print(region)
a=region.keys()
b=region.values()
```

```python
fig =plt.figure(figsize=(9,9))
plt.pie(x =b, labels = a)
plt.show
plt.legend()
```

```python
#type
type_count = df["Type"].value_counts()
type_count =dict(type_count)
print(type_count)
```

```python
j=list(type_count.keys())
v=list(type_count.values())
```

```python
fig =plt.figure(figsize=(5,5))
plt.bar(x = j, height=v)
plt.show()
```

```python
method_count =df["Method"].value_counts()
method_count =dict(method_count)
print(method_count)
```

```python
#scatterplot
price =df["Price"]
price =list(price)
distance=df["Distance"]
distance= list(distance)
fig =plt.figure(figsize=(15,15))
#line plot

plt.plot(distance,price,linestyle="",
         marker="o",
         markerfacecolor ="red",
         markeredgecolor="black")
plt.xlabel("distance")
plt.ylabel("price")
plt.title("distance vs price")
plt.show()
```
